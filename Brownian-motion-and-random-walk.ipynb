### Brownian motion and random walks

The purpose of this program is to simulate the motion of a particle in a gas or liquid as it is buffetted by random collisions with other molecules. The particle is confined to a L $\times$ L square grid.

```python
from random import randrange
from pylab import plot,show,xlim,ylim
import numpy as np
import math

L = 101
N = 20000

def walk1(N,L):
    
    # position represented by i,j = 0 ... L-1
    i = (L/2 - 1)
    j = (L/2 - 1)
    x = []
    y = []

    for k in range(0,N-1):
        dir = randrange(4)
        
        # moving particle randomly 
        # either left, right, up, or down for N steps      
    
        if (dir == 0) & (i < (L-1)):
            i += 1
        if (dir == 1) & (i > 0):
            i -= 1
        if (dir == 2) & (j < (L-1)):
            j += 1
        if (dir == 3) & (j > 0):
            j -= 1
            
        # storing the motion of the particle
        x.append(i)
        y.append(j)
    
    plot(x,y)
    xlim(0,(L-1))
    ylim(0,(L-1))
    show()

    
# plotting random motion for 5 particles
for n in range(5):
    print(n)
    walk1(N,L)

```

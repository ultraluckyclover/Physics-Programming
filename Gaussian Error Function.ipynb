### Using numerical integration for the Gaussian error function $ E(x) = \int_{0}^{x} e^{-t^2}dt $ using the trapezoidal rule and graphing $E(x)$ as a function of $x$.




```python
#! /usr/bin/env python3
import math
from pylab import plot,show

N = 30
a = 0.0
b = 3.0
w = (b-a)/N
xpoints = []
ypoints = []

def f(t):
    return math.exp(-(t**2)) #Gaussian error function

s = 0.5*f(a) + 0.5*f(b)

for i in range(0,N+1):
    if (i==0) or (i==N):
        xpoints.append(a+i*w)
        ypoints.append(f(a+i*w))
    else:
        xpoints.append(a+i*w)
        ypoints.append(f(a+i*w))
        s+= f(a+i*w)
        
plot(xpoints,ypoints)
show()
          
print(w*s)
```
